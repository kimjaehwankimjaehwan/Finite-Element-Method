##**현의 문제(Brachistochrone Problem)**

- 문제의 제시와 해답
문제 제시자: 요한 베르누이(Johann Bernoulli)가 1696년에 최초로 이 문제를 제시했습니다. 그는 당대의 수학자들에게 현의 문제를 해결해 보라고 도전장을 던졌습니다.

해답을 찾은 사람들:

  - 아이작 뉴턴(Isaac Newton)
  - 고트프리트 라이프니츠(Gottfried Leibniz)
  - 야코프 베르누이(Jakob Bernoulli, 요한 베르누이의 형제)
  - 로피탈(Marquis de L'Hôpital)
  - 이들은 모두 독립적으로 현의 문제를 해결하였고, 특히 아이작 뉴턴은 이 문제를 단 하루 만에 풀어냈다고 알려져 있습니다.

- 현의 문제의 해답
이들에 의해 최단 시간 경로가 사이클로이드 곡선임이 밝혀졌습니다.
뉴턴과 라이프니츠의 해법은 변분법의 초기 발전에 큰 기여를 하였으며, 변분법이 수학에서 중요한 도구로 자리 잡게 되는 계기가 되었습니다.


사이클로이드의 매개변수 방정식

사이클로이드 곡선의 매개변수 방정식은 다음과 같습니다:

<center>
$𝑥(𝜃)=𝑟(𝜃−sin⁡𝜃)$
</center>
<center>
$𝑦(𝜃)=𝑟(1−cos⁡𝜃)$
</center>

여기서:

$𝑟$은 생성하는 원의 반지름

$𝜃$는 매개변수로 각도를 나타냅니다.




```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 사이클로이드 곡선의 파라미터 설정
r = 1  # 반지름
theta = np.linspace(0, 2 * np.pi, 500)

# 사이클로이드 곡선의 x, y 좌표
x = r * (theta - np.sin(theta))
y = r * (1 - np.cos(theta))

# z축을 높이로 활용하여 3D 효과를 냄 (단순히 z축을 x 또는 y의 변형으로 활용)
z = np.sin(theta)

# 3D 시각화
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 사이클로이드 그리기
ax.plot(x, y, z, label='Cycloid Curve')

# 축 설정
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Brachistochrone Problem - Cycloid Curve in 3D')
ax.legend()

# 그래프 표시
plt.show()
```

# 1. 아이작 뉴턴 (Isaac Newton)의 방법
뉴턴은 변분법을 사용하지 않았지만, 역학적 원리를 이용하여 문제를 해결했습니다. 그는 다음과 같은 방식으로 접근했습니다.

* 뉴턴의 에너지 접근법:
중력가속도를
$g$라고 하고, 시작점과 도착점의 높이 차이를
$y$로 두었을 때, 물체의 운동 에너지와 위치 에너지 사이의 관계를 이용했습니다.
물체의 속력
$v$는 위치 에너지의 감소량에 의해 결정됩니다.
위치 에너지를 이용해 물체의 속력을 계산하면:
<center>

  $𝑣=2𝑔𝑦$
</center>
뉴턴은 이 속력을 이용하여 물체가 이동하는 데 걸리는 시간을 구했습니다. 이동 시간
$dt$는 다음과 같이 표현할 수 있습니다.
<center>
  $𝑑𝑡=\frac{𝑑𝑠}{𝑣}$
</center>

* 여기서 $ds$는 경로의 미소 길이입니다.

* 이를 이용해 이동 시간을 $𝑥$와 $𝑦$의 함수로 표현하면:
<center>
  $𝑇=∫\frac{𝑑𝑥^2+𝑑𝑦^2}{\sqrt{2𝑔𝑦}}$
</center>  
뉴턴은 이 식을 최소화하여 사이클로이드 곡선이 최적의 경로임을 확인했습니다.

```python
import numpy as np
import matplotlib.pyplot as plt

# Cycloid curve parameters
r = 1  # radius of the generating circle
t_vals = np.linspace(0, 4 * np.pi, 1000)  # parameter t values for the cycloid

# Cycloid parametric equations
x_cycloid = r * (t_vals - np.sin(t_vals))
y_cycloid = r * (1 - np.cos(t_vals))

def newton_cycloid():
    plt.figure(figsize=(8, 6))
    plt.plot(x_cycloid, y_cycloid, label="Cycloid (Newton's Approach)", color='blue')
    plt.title("Newton's Approach: Cycloid Curve")
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.grid(True)
    plt.legend()
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

newton_cycloid()
```

```python
def newton_cycloid_3d():
    r = 1  # Radius of the generating circle
    t_vals = np.linspace(0, 4 * np.pi, 1000)
    x_cycloid = r * (t_vals - np.sin(t_vals))
    y_cycloid = r * (1 - np.cos(t_vals))
    z_cycloid = t_vals  # Additional dimension for 3D

    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x_cycloid, y_cycloid, z_cycloid, label="Cycloid (Newton's Approach)", color='blue')
    ax.set_title("Newton's Approach: Cycloid Curve in 3D")
    ax.set_xlabel('x-axis')
    ax.set_ylabel('y-axis')
    ax.set_zlabel('z-axis')
    ax.legend()
    plt.show()

newton_cycloid_3d()

```

# 2. 고트프리트 라이프니츠 (Gottfried Leibniz)의 방법
라이프니츠는 미적분과 변분법을 사용하여 문제를 해결했습니다.

라이프니츠의 변분법 접근법:
경로의 이동 시간을 최소화하기 위해 다음과 같은 시간 함수
$T$를 고려했습니다.

<center>
$𝑇=∫\frac{\sqrt{1+(\frac{𝑑𝑦}{𝑑𝑥})^2}}{\sqrt{2𝑔𝑦}}𝑑𝑥$
</center>

라이프니츠는 미분방정식을 통해 이 식의 최소화 문제를 해결했습니다. 이 과정에서 경로가 사이클로이드가 되어야 한다는 결론을 도출했습니다.

변분법에서 오일러-라그랑주 방정식을 사용하면 다음 식이 성립합니다:
<center>
$\frac{𝑑}{𝑑𝑥}(\frac{∂𝐹}{∂𝑦′})−\frac{∂𝐹}{∂𝑦}=0$
</center>

여기서 $𝐹=\frac{\sqrt{1+(𝑦′)^2}}{\sqrt{𝑦}}$ 이며, 이를 풀어 사이클로이드 곡선을 얻을 수 있었습니다.

```python
def leibniz_cycloid():
    plt.figure(figsize=(8, 6))
    plt.plot(x_cycloid, y_cycloid, label="Cycloid (Leibniz's Approach)", color='red')
    plt.title("Leibniz's Approach: Cycloid Curve")
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.grid(True)
    plt.legend()
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

leibniz_cycloid()
```

```python
def leibniz_cycloid_3d():
    r = 1
    t_vals = np.linspace(0, 4 * np.pi, 1000)
    x_cycloid = r * (t_vals - np.sin(t_vals))
    y_cycloid = r * (1 - np.cos(t_vals))
    z_cycloid = t_vals

    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x_cycloid, y_cycloid, z_cycloid, label="Cycloid (Leibniz's Approach)", color='red')
    ax.set_title("Leibniz's Approach: Cycloid Curve in 3D")
    ax.set_xlabel('x-axis')
    ax.set_ylabel('y-axis')
    ax.set_zlabel('z-axis')
    ax.legend()
    plt.show()

leibniz_cycloid_3d()
```

# 3. 야코프 베르누이 (Jakob Bernoulli)의 방법
야코프 베르누이는 변분법을 공식적으로 사용하여 문제를 해결했습니다.

오일러-라그랑주 방정식을 이용한 방법:
야코프 베르누이는 함수
$y(x)$에 대해 이동 시간을 최소화하기 위해 오일러-라그랑주 방정식을 사용했습니다.
오일러-라그랑주 방정식은 다음과 같습니다.
<center>
$\frac{𝑑}{𝑑𝑥}(\frac{∂𝐿}{∂𝑦′})−\frac{∂𝐿}{∂𝑦}=0$
</center>

여기서
<center>
$𝐿=\frac{\sqrt{1+(\frac{𝑑𝑦}{𝑑𝑥})^2}}{\sqrt{2𝑔𝑦}}$
</center>

 입니다.

이 식을 풀어서 얻어지는 경로가 바로 사이클로이드 곡선이며, 이를 매개변수 방정식으로 표현하면 다음과 같습니다:
<center>
𝑥=𝑟(𝑡−sin⁡(𝑡))
</center>
<center>
𝑦=𝑟(1−cos⁡(𝑡))
</center>

여기서
$𝑟$은 주어진 원의 반지름입니다.


```python
def bernoulli_cycloid():
    plt.figure(figsize=(8, 6))
    plt.plot(x_cycloid, y_cycloid, label="Cycloid (Jakob Bernoulli's Approach)", color='green')
    plt.title("Jakob Bernoulli's Approach: Cycloid Curve")
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.grid(True)
    plt.legend()
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

bernoulli_cycloid()
```

```python
def bernoulli_cycloid_3d():
    r = 1
    t_vals = np.linspace(0, 4 * np.pi, 1000)
    x_cycloid = r * (t_vals - np.sin(t_vals))
    y_cycloid = r * (1 - np.cos(t_vals))
    z_cycloid = t_vals

    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot(x_cycloid, y_cycloid, z_cycloid, label="Cycloid (Jakob Bernoulli's Approach)", color='green')
    ax.set_title("Jakob Bernoulli's Approach: Cycloid Curve in 3D")
    ax.set_xlabel('x-axis')
    ax.set_ylabel('y-axis')
    ax.set_zlabel('z-axis')
    ax.legend()
    plt.show()

bernoulli_cycloid_3d()
```

# 4. 로피탈 (Marquis de L'Hôpital)의 방법
로피탈은 라이프니츠의 미적분 방식을 따랐고, 본질적으로 변분법을 사용하여 문제를 해결했습니다.

변분법을 통한 일반적 해법:
로피탈은 시간 함수
$𝑇$를 최소화하는 일반적인 변분 문제로 접근했습니다. 그 역시 다음 식을 최소화하려고 했습니다.

<center>
$𝑇=∫\frac{1+(\frac{𝑑𝑦}{𝑑𝑥})^2}{\sqrt{2𝑔𝑦} }𝑑𝑥$
</center>

그는 이를 해결하기 위해 오일러-라그랑주 방정식을 적용했고, 다른 수학자들과 마찬가지로 사이클로이드 곡선을 도출했습니다.

```python
def lhopital_cycloid():
    plt.figure(figsize=(8, 6))
    plt.plot(x_cycloid, y_cycloid, label="Cycloid (L'Hôpital's Approach)", color='purple')
    plt.title("L'Hôpital's Approach: Cycloid Curve")
    plt.xlabel('x-axis')
    plt.ylabel('y-axis')
    plt.grid(True)
    plt.legend()
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

lhopital_cycloid()
```
